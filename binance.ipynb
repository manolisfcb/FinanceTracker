```python
import requests
import base64
import os

def download_cvm_pdf_via_ajax(numero_protocolo, codigo_instituicao="1"):
    session = requests.Session()
    headers_base = {
        "User-Agent": "Mozilla/5.0",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "Referer": (
            "https://www.rad.cvm.gov.br/ENET/frmExibirArquivoIPEExterno.aspx"
            f"?NumeroProtocoloEntrega={numero_protocolo}"
        )
    }

    # 1) Hacemos GET para establecer la cookie de sesión
    url_page = "https://www.rad.cvm.gov.br/ENET/frmExibirArquivoIPEExterno.aspx"
    resp = session.get(url_page, params={"NumeroProtocoloEntrega": numero_protocolo},
                       headers=headers_base)
    resp.raise_for_status()

    # 2) Preparamos el payload JSON
    payload = {
        "codigoInstituicao": str(codigo_instituicao),
        "numeroProtocolo": str(numero_protocolo),
        "token": "",
        "versaoCaptcha": ""
    }

    # 3) Cabeceras para la llamada AJAX
    ajax_headers = {
        "User-Agent": headers_base["User-Agent"],
        "Accept": "application/json, text/javascript, */*; q=0.01",
        "Content-Type": "application/json; charset=UTF-8",
        "X-Requested-With": "XMLHttpRequest",
        "Referer": headers_base["Referer"]
    }

    # 4) POST al método que devuelve el PDF (en base64 dentro de JSON)
    url_ajax = url_page + "/ExibirPDF"
    r2 = session.post(url_ajax, json=payload, headers=ajax_headers)
    r2.raise_for_status()

    # 5) La respuesta es JSON, contiene el PDF codificado en base64, por ejemplo:
    #    { "d": "JVBERi0xLjQKJeLjz9MNCjMgMCBvYmoKPD..." }
    data = r2.json()
    # dependiendo de la API puede venir en data['d'] o data['pdfBase64']
    b64 = data.get("d") or data.get("pdfBase64")
    if not b64:
        raise RuntimeError("No se encontró el PDF en la respuesta JSON.")

    # 6) Decodificar y guardar
    pdf_bytes = base64.b64decode(b64)
    filename = f"cvm_{numero_protocolo}.pdf"
    with open(filename, "wb") as f:
        f.write(pdf_bytes)

    print(f"PDF descargado en {os.path.abspath(filename)}")
    return filename

if __name__ == "__main__":
    download_cvm_pdf_via_ajax("1363054")
```
